```python
import matplotlib.pyplot as plt

files = {
    "easy_train.txt": "green",
    "easy_test.txt": "green",
    "medium_train.txt": "blue",
    "medium_test.txt": "blue",
    "hard_train.txt": "red",
    "hard_test.txt": "red"
}


for file, color in files.items():
    x_vals = []
    y_vals = []
    with open(file, 'r') as f:
        for line in f:
            x_str, y_str = line.strip().split(" ")
            x_vals.append(float(x_str))
            y_vals.append(float(y_str))

    plt.figure(figsize=(6, 4))
    plt.scatter(x_vals, y_vals, alpha=0.6, s=10, color=color)
    plt.title(f"Scatter Plot - {file}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
